```python
mv_2003['기준일'] = pd.to_datetime(mv_2003['기준일'],format='%Y-%m-%d')
```

```python
mv_preprocess = mv_2003.loc[mv_2003.groupby(['영화명'])['기준일'].idxmax()]
mv_preprocess
```

```python
country_counts = mv_preprocess.groupby('대표국적')['대표국적'].count()

colors = ['#40E0D0', '#ffc000', '#56B4E9', '#FFFFF0', '#FFEBCD', '#D2691E', '#FFFAF0', '#8fd9b6', '#ff9999']

plt.figure(figsize=(7,7))

plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', colors=colors)
plt.axis('equal')
plt.title('대표국적 비율')
plt.show()
```

# 파이차트 추출2 ( Others 묶기 )

```python
country_counts_others = mv_preprocess.groupby('대표국적')['대표국적'].count().reset_index(name='영화개수')
country_counts_others
```

```python
country_counts_others[country_counts_others.대표국적 == '미국']
```

```python
country_counts_others[country_counts_others.영화개수 >= 100]
```

```python
others_sum = country_counts_others[country_counts_others['영화개수'] < 100]['영화개수'].sum()
others_sum
```

```python
coco = country_counts_others[country_counts_others.영화개수 >= 300]
coco.set_index('대표국적', inplace=True)
coco = coco.sort_values(by=coco.columns[0],ascending=False)
coco.loc['others'] = [others_sum]
coco
```

```python
coco.index
```

```python
coco.values.flatten()
```

```python
labels = coco.index.tolist()
ratio = coco.values.flatten().tolist()
colors = ['#40E0D0', '#ffc000', '#56B4E9', '#FFFFF0', '#FFEBCD', '#D2691E', '#FFFAF0', '#8fd9b6', '#ff9999']

sorted_ratio, sorted_labels = zip(*sorted(zip(ratio, labels), reverse=True))

plt.figure(figsize=(10,10))


plt.pie(sorted_ratio, labels=sorted_labels, autopct='%.1f%%', startangle=90, radius=0.9, textprops={'fontsize':14}, colors=colors)
plt.title('영화 대표국적 비율', fontsize=20)
plt.ylabel('')

plt.show()
```
