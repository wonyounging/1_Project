```python
# 주소 -> 위도, 경도

def get_location(address):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
  # 'KaKaoAK '는 그대로 두시고 개인키만 지우고 입력해 주세요.
  # ex) KakaoAK 6af8d4826f0e56c54bc794fa8a294
  headers = {"Authorization": "KakaoAK 64e44d0d352850325f1af2de2780c30d"}
  api_json = json.loads(str(requests.get(url,headers=headers).text))
  address = api_json['documents'][0]['address']
  crd = {str(address['y']), str(address['x'])}
  address_name = address['address_name']

  return crd

crd = get_location("서울특별시 구로구 신도림동 692 번지 디큐브시티 제타워동 7,8층 일부")
print(crd)
```

```python
list_lat_571=[]
list_lng_571=[]
for set1 in location:
    one_iter = set1.__iter__()
    temp1 = float(one_iter.__next__())
    temp2 = float(one_iter.__next__())
    if temp1 < 100:
        list_lat_571.append(temp1)
    else :
        list_lng_571.append(temp1)
    if temp2 < 100:
        list_lat_571.append(temp2)
    else :
        list_lng_571.append(temp2)
        
print(list_lat_571[:5])
print(list_lng_571[:5])
```

```python
theater2 = theater.reset_index(drop=True) # 인덱스값 초기화
theater2 = theater2.drop(index=list1, axis = 0)
theater2.reset_index(drop=True, inplace=True)
theater2
```

```python
df_lat = pd.DataFrame({'위도' : list_lat_571})
theater2 = pd.concat([theater2,df_lat], axis=1)
df_lng = pd.DataFrame({'경도' : list_lng_571})
df_theater2 = pd.concat([theater2,df_lng], axis=1)
df_theater2
#theater2.merge(df_lat, on=)
```
