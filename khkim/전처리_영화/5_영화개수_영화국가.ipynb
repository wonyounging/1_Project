```python
mv_2017['기준일'] = pd.to_datetime(mv_2017['기준일'],format='%Y-%m-%d')
```

```python
mv_preprocess = mv_2017.loc[mv_2017.groupby(['영화명'])['기준일'].idxmax()]
mv_preprocess
```

```python
country_counts = mv_preprocess.groupby('대표국적')['대표국적'].count()

plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('대표국적 비율')
plt.show()
```
