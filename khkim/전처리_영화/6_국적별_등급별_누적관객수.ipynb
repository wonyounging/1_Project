# 2. 국적별 누적관객수

```python
mv_2018['기준일'] = pd.to_datetime(mv_2018['기준일'],format='%Y-%m-%d')
```

```python
mv_lately_order = mv_2018.loc[mv_2018.groupby(['영화명'])['기준일'].idxmax()]
mv_lately_order.head()
```

```python
mv_last_date = mv_2018.loc[mv_2018.groupby(['영화명'])['기준일'].idxmax()]
mv_last_date.head()
```

```python
mv_country_moviegoer = mv_last_date.groupby('대표국적')['누적관객수'].sum().reset_index(name='국적별누적관객수')
mv_country_moviegoer = mv_country_moviegoer.sort_values(by='국적별누적관객수', ascending=False)[:10]
mv_country_moviegoer
```

```python
mv_country_moviegoer_div = mv_country_moviegoer[:]
mv_country_moviegoer_div['국적별누적관객수'] = mv_country_moviegoer_div['국적별누적관객수']/10000
mv_country_moviegoer_div.plot.bar(x='대표국적', y='국적별누적관객수', rot=0, figsize=(10, 5), fontsize=10)
plt.title('국적별 누적관객수')
plt.ylabel('국적별누적관객수/10000')
plt.show()
```

# 3. 등급별 누적관객수

```python
mv_last_date.groupby('등급')['누적관객수'].sum()
```

```python
mv_last_date.groupby('등급')['누적관객수'].sum().reset_index()
```

```python
mv_last_date.groupby('등급')['누적관객수'].sum().reset_index(name='등급별누적관객수')
```

```python
mv_rate_moviegoer = mv_last_date.groupby('등급')['누적관객수'].sum().reset_index(name='등급별누적관객수')
mv_rate_moviegoer = mv_rate_moviegoer.sort_values(by='등급별누적관객수', ascending=False)[:4]

mv_rate_moviegoer_div = mv_rate_moviegoer[:]
mv_rate_moviegoer_div['등급별누적관객수'] = mv_rate_moviegoer_div['등급별누적관객수']/10000
mv_rate_moviegoer_div.plot.bar(x='등급', y='등급별누적관객수', rot=0, figsize=(10, 5), fontsize=10)
plt.title('등급별 누적관객수')
plt.ylabel('등급별누적관객수/10000')
plt.show()
```
