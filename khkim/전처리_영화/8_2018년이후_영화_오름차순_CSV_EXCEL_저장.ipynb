# 데이터 CSV, EXCEL 저장

```python
mv_2017_csv_2 = mv_2017_csv_2[mv_2017_csv_2['기준연도'] > 2017]
df_order = mv_2017_csv_2.sort_values(by=['기준일', '순위'])
df_order = df_order.reset_index(drop=True)
df_order.to_csv('./resources/완_기간별_ordered_180101_to_230509_ver3.csv', index=False, encoding='UTF-8')
df_order.to_excel('./resources/완_기간별_ordered_180101_to_230509_ver3.xlsx', index=False)
```
