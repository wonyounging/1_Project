# 개봉일이 2017년 1월 1일 이후인 영화 추출

```python
file_movie_range_7years = './resources/기간별_일별_170108_to_230508_수정.csv'
df_mv_7years = pd.read_csv(file_movie_range_7years)
df_mv_7years
```

```python
df_mv_7years['기준일'] = pd.to_datetime(df_mv_7years['기준일'],format='%Y-%m-%d')
```

```python
df_2017 = df_mv_7years[df_mv_7years['개봉일'] >= '2017-01-01']
df_2017
```

```python
df_2017.to_csv('./resources/기간별_일별_170101_to_230509_ver1.csv', index=False)
```

# 영화명 중복 체크

```python
df_2017.head()
```

```python
df_n17 = df_2017.loc[df_2017.groupby('영화명')['개봉일'].count()]
```

```python
df_2017.columns
```

```python
# 중복 영화 제거
df_unique_movies = df_2017.drop_duplicates(subset=['영화명'], keep=False)

# 중복 영화를 다시 선택하고, 영화명과 개봉일로 그룹화하여 개수 세기
df_duplicate_movies = df_2017[df_2017.duplicated(subset=['영화명'], keep=False)]
df_duplicate_movie_counts = df_duplicate_movies.groupby(['영화명', '개봉일']).size().reset_index(name='개수')

# 개수가 2 이상인 경우만 선택하여 출력
df_duplicate_movie_counts[df_duplicate_movie_counts['개수'] >= 2]
```
