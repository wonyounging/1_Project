> 배우 추출 함수만들기

```python
def actors_count(mv_genre):
    df_actors = mv_genre['배우'].apply(lambda x : str(x).split(','))
    df_actors_explode = df_actors.explode()
    df_keyword = pd.DataFrame({'keyword':df_actors_explode})
    df_keyword['count'] = df_keyword['keyword'].str.len()
    df_keyword_sorted = df_keyword.groupby('keyword', as_index = False).count().sort_values('count', ascending = False)
    df_keyword_sorted_2 = df_keyword_sorted.iloc[1:].reset_index().drop('index', axis=1)
    return df_keyword_sorted_2
```

### order, groupby해서 '배우'로 묶어서 dataframe 추출

```python
def get_actor_list(movies):
    movies['기준일'] = pd.to_datetime(movies['기준일'],format='%Y-%m-%d')
    movies['개봉일'] = pd.to_datetime(movies['개봉일'],format='%Y-%m-%d')
    movies_v1 = movies.sort_values(by=['기준일','순위'])
    movies_v2 = movies_v1.loc[movies_v1.groupby(['영화명','개봉일'])['기준일'].idxmax()]
    mv_2003_million = movies_v2[movies_v2['누적관객수'] > 10**7] # 10**6 은 백만, 10**7은 천만
    df_actors_v2 = mv_2003_million[:]
    df_actors_v2['배우List'] = df_actors_v2['배우'].apply(lambda x : str(x).split(','))
    df_actors_explode_v2 = df_actors_v2.explode(column=['배우List'],ignore_index=True)
    df_group_actor_movie = df_actors_explode_v2.groupby('배우List')['영화명'].apply(list).reset_index(name='영화목록')
    df_group_actor_movie['영화촬영개수'] = df_group_actor_movie['영화목록'].apply(lambda x : len(x))
    df_gp_actor = df_group_actor_movie.sort_values('영화촬영개수', ascending = False)
    return df_gp_actor
```

```python
10**7
```

```python
df_movies_action = mv_2003[mv_2003['장르'].str.contains('액션')]
df_action = get_actor_list(df_movies_action)
df_action
```

```python
df_action.to_csv('./resources/mv_action_10million_actors_ver6.csv', index=False, encoding='UTF-8')
```

```python
df_movies_adventure = mv_2003[mv_2003['장르'].str.contains('어드벤처')]
df_adventure = get_actor_list(df_movies_adventure)
df_adventure
```

```python
df_adventure.to_csv('./resources/mv_adventure_10million_actors_ver6.csv', index=False, encoding='UTF-8')
```
