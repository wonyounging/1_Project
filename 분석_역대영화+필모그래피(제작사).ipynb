```python
filmo_data = pd.read_excel('./use_data/필모그래피_통합.xlsx', engine='openpyxl')
filmo_data
```

# 제작사 필터링

```python
producer_datas = filmo_data[filmo_data['분야'].str.contains('제작사')]
producer_datas
```

# 사용 컬럼 선택

```python
producer_data = producer_datas[['배급사', '영화명', '공식통계 매출액','KOBIS통계 매출액']]
producer_data
```

# 결측값 대체

```python
for i in range(len(producer_data)):
    if producer_data.iloc[i,2] > 0:
        producer_data.iloc[[i],[2]] = producer_data.iloc[[i],[3]]
producer_data
```

# 결측값 확인, 제거

```python
# KOBIS통계 매출액 공식 통계 매출액으로 대체
producer_data[producer_data['공식통계 매출액'].isnull()]
```

```python
producer_data = producer_data.dropna(axis = 0)
producer_data
```

```python
# 매출액이 0인 행 삭제
producer_data[producer_data['공식통계 매출액'] == 0]
```

```python
# 현재 데이터프레임 확인
producer_data = producer_data.drop(producer_data[producer_data['공식통계 매출액'] == 0].index)
producer_data
```

# 필요 없는 컬럼 삭제

```python
# KOBIS 통계 매출액 컬럼 삭제
producer_data = producer_data.drop(columns='KOBIS통계 매출액')
producer_data
```

```python
producer_data[producer_data['배급사'] == '(주)대명문화공장']
```

# 제작사별 단편 최고 매출액, 영화명 추출
## (기본 프레임)

```python
bb = producer_data.loc[producer_data.groupby(['배급사'])['공식통계 매출액'].idxmax()]
bb
```

# 제작사별 제작 작품수 추출, 병합

```python
# 제작사별 제작 작품수
tot_cnt = pd.DataFrame(producer_data.groupby("배급사")['공식통계 매출액'].count())
tot_cnt = tot_cnt.rename(columns={'공식통계 매출액': '제작 작품수'})
tot_cnt = tot_cnt.reset_index()
tot_cnt
```

```python
bb = bb.merge(tot_cnt, on='배급사', how='left')
bb
```

# 제작사별 총 매출 추출, 병합

```python
# 제작사별 총 매출
tot_sum = pd.DataFrame(producer_data.groupby("배급사")['공식통계 매출액'].sum())
tot_sum = tot_sum.rename(columns={'공식통계 매출액': '총 매출액'})
tot_sum = tot_sum.reset_index()
tot_sum
```

```python
bb = bb.merge(tot_sum, on='배급사', how='left')
bb
```

# 제작사별 매출 평균 추출, 병합

```python
# 제작사별 매출 평균
tot_avg = pd.DataFrame(producer_data.groupby("배급사")['공식통계 매출액'].mean())
tot_avg = tot_avg.rename(columns={'공식통계 매출액': '매출액 평균'})
tot_avg = tot_avg.reset_index()
tot_avg
```

```python
bb = bb.merge(tot_avg, on='배급사', how='left')
bb
```

# 컬럼 순서 변경

```python
bb = bb[['배급사', '제작 작품수', '영화명', '공식통계 매출액', '총 매출액', '매출액 평균']]
bb
```

# 컬럼명 변경

```python
bb = bb.rename(columns={'배급사':'제작사', '공식통계 매출액':'단편 최고 매출액'})
bb
```

# 제작사별 총 매출 대비 단편 최고 매출 비율 컬럼 추가

```python
bb['총 매출 대비 단편 최고 매출 비율'] = ''
bb
```

```python
# 비율 계산
bb['총 매출 대비 단편 최고 매출 비율'] = (bb['단편 최고 매출액'] / bb['총 매출액'] * 100).round(2)
bb
```

```python
bb.to_csv('./제작사.csv', index=False)
```
