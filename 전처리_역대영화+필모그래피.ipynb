# 역대 영화의 배급사 검색

```python
movie_200_info_data
```

```python
a = movie_200_info_data[['영화명', '배급사', '개봉일', '매출액', '관객수']]
a
```

```python
b = filmo_data[['영화명', '배급사', '제작연도', '장르']]
b
```

```python
a.set_index('영화명', inplace=True)
b.set_index('영화명', inplace=True)
```

```python
result = a.join(b, lsuffix='_a', rsuffix='_b')
result
```

```python
result = result.reset_index()
result
```

```python
# 배급사_a와 배급사_b가 다른 항목 제거(역대 배급사 기준)
result = result.drop_duplicates(['영화명'])
```

```python
result
```

```python
result[result['배급사_a'] != result['배급사_b']]
```

```python
# 배급사_b 컬럼 제거
result.drop('배급사_b', axis=1, inplace=True)
```

```python
# 배급사_a 컬럼 배급사로 변경
result.rename(columns={'배급사_a' : '배급사'}, inplace=True)
```

```python
result
```

```python
result.to_csv('./역대영화+필모그래피.csv', index=False)
```
