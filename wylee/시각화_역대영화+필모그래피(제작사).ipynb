```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
dist_data = pd.read_csv('./제작사.csv')
dist_data
```

```python
dist_data = dist_data.sort_values(by=['총 매출액'], ascending=False)
dist_data
```

```python
# 그래프 크기 설정
plt.figure(figsize=(15, 10))

# 총 매출액에 대한 막대 그래프
plt.subplot(1, 3, 1)
plt.bar(dist_data['제작사'], dist_data['총 매출액'])
plt.bar(dist_data['제작사'], dist_data['단편 최고 매출액'], label='단편 최고 매출액', color='red')
plt.xlabel('제작사')
plt.ylabel('총 매출액 (억)')
plt.title('총 매출액')
plt.xticks(rotation=45, ha='right')  # x축 값 회전

formatter = ticker.FuncFormatter(lambda x, pos: '{:.0f} 천억'.format(x / 1e11))
plt.gca().yaxis.set_major_formatter(formatter)

# 매출액 평균에 대한 막대 그래프
plt.subplot(1, 3, 2)
plt.bar(dist_data['제작사'], dist_data['매출액 평균'])
plt.xlabel('제작사')
plt.ylabel('매출액 평균 (억)')
plt.title('매출액 평균')
plt.xticks(rotation=45, ha='right')  # x축 값 회전

formatter = ticker.FuncFormatter(lambda x, pos: '{:.0f} 억'.format(x / 1e8))
plt.gca().yaxis.set_major_formatter(formatter)

# 단편 최고 매출액에 대한 막대 그래프
plt.subplot(1, 3, 3)
plt.bar(dist_data['제작사'], dist_data['단편 최고 매출액'], color='red')
plt.xlabel('제작사')
plt.ylabel('단편 최고 매출액 (백만)')
plt.title('단편 최고 매출액')
plt.xticks(rotation=45, ha='right')  # x축 값 회전

# y축의 단위를 억으로 설정
formatter = ticker.FuncFormatter(lambda x, pos: '{:.0f} 억'.format(x / 1e8))
plt.gca().yaxis.set_major_formatter(formatter)

# 서브플롯 간 간격 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```
