# 모든 데이터 불러오고 합치기

```python
import glob
```

```python
path = './data1/*'
file_list = glob.glob(path)
[file for file in file_list if file.endswith('.xlsx')]
```

```python
len(file_list)
```

```python
# data = pd.DataFrame(colums = ['지역', ' 년도', '월','한국영화 상영편수', '한국영화 매출액',
#         '한국영화 관객수', '한국영화 점유율', '외국영화 상영편수', '외국영화 매출액',
#         '외국영화 관객수', '외국영화 점유율', '전체 상영편수', '전체 매출액',
#         '전체 관객수', '전체 점유율'])
# data

data = pd.DataFrame()
data
```

```python
for i in range(len(file_list)):
    print(file_list[i])
    add = pd.read_excel(file_list[i], engine='openpyxl')

    ym = add.loc[2][0]
    year = ym.split(' ')[5].split('~')[0].split('-')[0]
    month = ym.split(' ')[5].split('~')[0].split('-')[1]
    print(year, month)
    
    add = pd.read_excel(file_list[i], engine='openpyxl', header = 4)
    add.insert(1, '년도', year)
    add.insert(2, '월', month)
    
    add = add.rename(columns={'지역':'지역', '년도':'년도', '월':'월',
                              '한국영화':'한국영화 상영편수', 'Unnamed: 2':'한국영화 매출액',
                              'Unnamed: 3':'한국영화 관객수', 'Unnamed: 4':'한국영화 점유율',
                              '외국영화':'외국영화 상영편수', 'Unnamed: 6':'외국영화 매출액',
                              'Unnamed: 7':'외국영화 관객수', 'Unnamed: 8':'외국영화 점유율',
                              '전체':'전체 상영편수', 'Unnamed: 10':'전체 매출액',
                              'Unnamed: 11':'전체 관객수', 'Unnamed: 12':'전체 점유율'})
    
    add.drop([0], axis=0, inplace=True)
    add.drop([18], axis=0, inplace=True)
    
    data = pd.concat([data, add])
data
```
